# DataFrame

```python
lancers = pd.DataFrame({
    "des1" : np.random.randint(1, 7, size=(30)),
    "des2" : np.random.randint(1, 7, size=(30)),
    "des3" : np.random.randint(1, 7, size=(30)),
    "des4" : np.random.randint(1, 7, size=(30)),
    }
)
lancers.head()
```

```python
lancers.columns
```

```python
lancers.index
```

```python
type(lancers.des1)
```

```python
lancers["des1"] == lancers.des1
```

```python
lancers["des1"] is lancers.des1
```

```python
id(lancers.des1)
```

```python
id(lancers["des1"])
```

```python
lancers["Somme"] = lancers.des1 + lancers.des2 + lancers.des3 + lancers.des4
```

```python
lancers.tail()
```

```python
lancers.info()
```

```python
lancers[["des1", "des2", "des3", "des4"]]
```

```python
lancers.drop("Somme", axis=1)
```

```python
lancers["egalite"] = (lancers.drop("Somme", axis=1)
                            .drop("egalite", axis=1)
                            .apply(lambda x: len(set(x)) < 4, axis=1))
```
