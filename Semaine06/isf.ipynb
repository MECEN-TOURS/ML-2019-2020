```python
isf["nombre de redevables"] = isf["nombre de redevables"].apply(int)
isf["patrimoine moyen en €"] = isf["patrimoine moyen en €"].apply(int)
isf["impôt moyen en €"] = isf["impôt moyen en €"].apply(int)
```

```python
isf.dtypes
```

```python
isf.head()
```

```python
isf["Région"] = isf.index
isf.index = isf["Commune"]
```

```python
isf["total patricmoine"] = isf["nombre de redevables"] * isf["patrimoine moyen en €"]
```

```python
isf.head()
```

```python
intermediaire = isf.groupby("Départements").agg("sum")
intermediaire.head()
```

```python
resultat = intermediaire["total patricmoine"] / intermediaire["nombre de redevables"]
```

```python
patrimoine_dpt = resultat.astype(int)
```
