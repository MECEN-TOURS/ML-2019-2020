```python
phrase = "Le renard saute au dessus du chient fainéant."
mon_dictionnaire = dict()
for car in phrase:
    mon_dictionnaire[car] = mon_dictionnaire.get(car, 0) + 1
    
print(mon_dictionnaire)
```

```python
with open("frequence.csv", "w", encoding="utf8") as fichier:
    ecrivain = csv.writer(fichier)
    ecrivain.writerow(mon_dictionnaire)
```

**On voit que writerow itére sur le conteneur passé et écrit chaque élément séparé par une virgule.**

```python
help(csv.DictWriter)
```

```python
with open("frequence.csv", "w", encoding="utf8") as fichier:
    ecrivain = csv.DictWriter(fichier, mon_dictionnaire.keys())
    ecrivain.writeheader()
    ecrivain.writerow(mon_dictionnaire)
```

```python
def genere_frequence(entree):
    mon_dictionnaire = dict()
    for clef in entree:
        mon_dictionnaire[clef] = mon_dictionnaire.get(clef, 0) + 1

    return mon_dictionnaire
```

```python
phrase = "Le renard saute au dessus du chient fainéant."
deuxieme_phrase = "Une phrase bateau."
freq1 = genere_frequence(phrase)
freq2 = genere_frequence(deuxieme_phrase)
print(freq1)
print(freq2)
```

```python
clef1 = set(freq1.keys())
clef2 = set(freq2.keys())
```

```python
print(*dir(clef1))
```

```python
clefs = clef1.union(clef2)
```

```python
with open("frequence_bis.csv", "w", encoding="utf8") as fichier:
    ecrivain = csv.DictWriter(fichier, clefs)
    ecrivain.writeheader()
    ecrivain.writerow(freq1)
    ecrivain.writerow(freq2)
```
